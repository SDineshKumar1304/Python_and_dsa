### Second Piece (k, l) xl2 <= k <= xr2, yl2 <= l <= yr2

### Regions

**region 1**
(xl1,yl1) to (xr1,yr1) --> (1,1) to (2,2)

**region 2**
(xl2,yr2) to (xr2,yr2) --> (3,3) to (4,4)


- We have two rectangular regions on a chessboard, we need to place two pieces
- The two pieces are not on the same square.
- Both pieces are on squares of the same color.



###  Chessboard Layout

|    | 1    | 2    | 3    | 4    |
|----|------|------|------|------|
| 1  | (1,1) W | (1,2) B | (1,3) W | (1,4) B |
| 2  | (2,1) B | (2,2) W | (2,3) B | (2,4) W |
| 3  | (3,1) W | (3,2) B | (3,3) W | (3,4) B |
| 4  | (4,1) B | (4,2) W | (4,3) B | (4,4) W |



### Regions and Conditions

We are given two regions of squares:

- **Region 1 (Piece 1)** is defined by the coordinates (1,1) to (2,2). It includes the squares:
  - (1,1) White
  - (1,2) Black
  - (2,1) Black
  - (2,2) White

- **Region 2 (Piece 2)** is defined by the coordinates (3,3) to (4,4). It includes the squares:
  - (3,3) White
  - (3,4) Black
  - (4,3) Black
  - (4,4) White

### Matching Squares from Region 1 and Region 2



We need to check all pairs of squares `(i, j)` from Region 1 and `(k, l)` from Region 2 to satisfy two conditions:

1. **The squares must not be the same square**: `(i, j) != (k, l)`.
2. **The squares must have the same color**: `(i + j) % 2 == (k + l) % 2`.

### Step-by-Step Matching

#### For (1, 1) (White in Region 1):

- Compare with (3, 3) (White in Region 2):  
  `(1 + 1) % 2 == (3 + 3) % 2`: Both are White, so they match.

- Compare with (3, 4) (Black in Region 2):  
  `(1 + 1) % 2 != (3 + 4) % 2`: White and Black do not match.

- Compare with (4, 3) (Black in Region 2):  
  `(1 + 1) % 2 != (4 + 3) % 2`: White and Black do not match.

- Compare with (4, 4) (White in Region 2):  
  `(1 + 1) % 2 == (4 + 4) % 2`: Both are White, so they match.

**Valid matches for (1, 1)**: (3, 3) and (4, 4).

**count**  - 2

#### For (1, 2) (Black in Region 1):

- Compare with (3, 3) (White in Region 2):  
  `(1 + 2) % 2 != (3 + 3) % 2`: Black and White do not match.

- Compare with (3, 4) (Black in Region 2):  
  `(1 + 2) % 2 == (3 + 4) % 2`: Both are Black, so they match.

- Compare with (4, 3) (Black in Region 2):  
  `(1 + 2) % 2 == (4 + 3) % 2`: Both are Black, so they match.

- Compare with (4, 4) (White in Region 2):  
  `(1 + 2) % 2 != (4 + 4) % 2`: Black and White do not match.

**Valid matches for (1, 2)**: (3, 4) and (4, 3).

**count**  - 2

#### For (2, 1) (Black in Region 1):

- Compare with (3, 3) (White in Region 2):  
  `(2 + 1) % 2 != (3 + 3) % 2`: Black and White do not match.

- Compare with (3, 4) (Black in Region 2):  
  `(2 + 1) % 2 == (3 + 4) % 2`: Both are Black, so they match.

- Compare with (4, 3) (Black in Region 2):

```python
# Recursive 
'''
Recursive functions are functions that call themselves to solve a problem by breaking it into smaller, simpler subproblems.
 Recursive Functions have base case to stop the loop
'''
def  fact(n):
    return 1 if n==1 or n==0 else n*(fact(n-1)) # recursive call of factorial
     # n = 5  (n==1 or n==0) false --> 5*(fact(4)) = 5*(4*(fact(3))) --> 5*(4*(3(fact(2)))) --> 5*(4*(3*(2*(fact(1))))) --> 
     # --> 5*4*3*2*1 = 120
     # op --> 129
print(fact(5))
```

```python
# Sum
def sum_list(lst):
    if not lst:  # Base case: empty list
        return 0
    return lst[0] + sum_list(lst[1:])  # Recursive call

print(sum_list([1, 2, 3, 4]))  # Output: 10
```

```python
# Reversing String
def reverse_string(s):
    if len(s) == 0:  # Base case: empty string
        return ""
    return s[-1] + reverse_string(s[:-1])  # Recursive call

print(reverse_string("Python"))  # Output: "Python"
```

```python
def countdown(n):
    if n <= 0:  # Base case
        print("Blast off!")
        return
    print(n)
    countdown(n - 1)  # Recursive call

countdown(5)
# Output:
# 5
# 4
# 3
# 2
# 1
# Blast off!
```

```python
def gcd(a, b):
    if b == 0:  # Base case
        return a
    return gcd(b, a % b)  # Recursive call

print(gcd(48, 18))  # Output: 6
```

```python
# Stack using list

# Stack contains push,pop,display,size operations

# This is Last in First Out 
# Right Insertion
class Stack:
    def __init__(self):
        self.stack = []
        
    def display(self):
        if len(self.stack) == 0:
            print("Stack is empty")
        else:
            print(self.stack)
            
    def size(self):
        print("size :",len(self.stack))
        
    def push(self,data):
        self.stack.append(data)
        print(f"{data} is added to stack")
        
    def pop(self):
        if len(self.stack)==0:
            print("Stack is emty")
        else:
            self.stack.pop()
            print("Element removed")
    
s1 = Stack()
s1.display()
s1.push(10)
s1.push(20)
s1.push(30)
s1.display()
s1.size()
s1.pop()
s1.display()
s1.pop()
s1.display()
s1.pop()
s1.display()
```

```python
# left Insertion
# here we use appendleft and popleft
from collections import deque

class stack:
    def __init__(self):
        self.stack = deque()
    
    def display(self):
        if len(self.stack)==0:
            print("stack is Empty")
        else:
            print(self.stack)
            
    def push(self,data):
        self.stack.appendleft(data)
        print(f"{data} is added")
        
    def pop(self):
        if len(self.stack)==0:
            print("Stack is empty")
        else:
            self.stack.popleft()
            print("element removed")
    
s1 = stack()
s1.display()
s1.push(10)
s1.push(20)
s1.push(30)
s1.display()
s1.pop()
s1.display()
s1.pop()
s1.display()
s1.pop()
s1.display()
s1.pop()
```

# Stack with LifoQueue

```python
import queue

class Stack:
    def __init__(self):
        self.stack = queue.LifoQueue()
    
    def display(self):
        if self.stack.qsize() == 0:
            print("Stack is Empty")
        else:
            print(self.stack.queue)  # Display elements in the stack as a list
            
    def push(self, data):
        self.stack.put(data)
        print(f"{data} is added")
        
    def pop(self):
        if self.stack.qsize() == 0:
            print("Stack is empty")
        else:
            removed = self.stack.get()
            print(f"Element {removed} removed")
    
s1 = Stack()
s1.display()
s1.push(10)
s1.push(20)
s1.push(30)
s1.display()
s1.pop()
s1.display()
s1.pop()
s1.display()
s1.pop()
s1.display()
```

# Sorting

# bubble sort

```python
class BubbleSort:
    def __init__(self,a):
        self.a = a
    
    def bubblesort(self,a):
        for i in range(len(self.a)):
            for j in range(len(self.a)-1):
                if self.a[i]<self.a[j]:
                    self.a[i],self.a[j] = self.a[j],self.a[i]
        return self.a

a = [21,3,1,3,0]
b1 = BubbleSort(a)

b1.bubblesort(a)
```

# Insertion Sort

```python
class InsertionSort:
    def __init__(self,a):
        self.a = a
        
    def insertionsort(self,a):
        for i in range(1,len(self.a)):
            j = i
            while self.a[j-1]>self.a[j] and j>0:
                self.a[j],self.a[j-1] = self.a[j-1] ,self.a[j]
                j-=1
        return self.a
    
a = [21,3,1,3,0]
i1 = InsertionSort(a)

i1.insertionsort(a)
```

# Selection Sort

```python
class SelectionSort:
    def __init__(self,a):
        self.a = a
    
    def selectionsort(self,a):
        for i in range(len(self.a)):
            m = i
            for j in range(i+1,len(self.a)):
                if self.a[m]>self.a[j]:
                    m = j
            self.a[m],self.a[i] = self.a[i],self.a[m]
        return self.a
    
a = [21,3,1,3,0]
S1 = SelectionSort(a)

S1.selectionsort(a)
```

# Quick Sort

```python
import random

class QuickSort:
    def __init__(self,a):
        self.a = a
        
    
    def quicksort(self,a):
        if len(a)<=1:
            return a
        
        p = random.choice(a)
        
        l = [i for i in a if i<p]
        r = [i for i in a if i>p]
        m = [i for i in a if i==p]
        
        return self.quicksort(l)+m+self.quicksort(r)

a = [21,3,1,3,0]
Q1 = QuickSort(a)

Q1.quicksort(a)     
```

# merge Sort

```python
class MergeSort:
    def __init__(self,a):
        self.a = a
        
    def mergesort(self,a):
        if len(a)>1:
            l = a[:len(a)//2]
            r = a[len(a)//2:]

            self.mergesort(l)
            self.mergesort(r)

            a.clear()

            while len(l)>0 and len(r)>0:
                if l[0]<r[0]:
                    a.append(l.pop(0))
                else:
                    a.append(r.pop(0))

            a.extend(l)
            a.extend(r)
        return a
    
a= [21,3,1,3,0]
m1= MergeSort(a)
m1.mergesort(a)       
```

# Linear Search

```python
class LinearSearch:
    def __init__(self,a,t):
        self.a = a
        self.t = t
        
    def linearsearch(self,a,t):
        for i in range(len(self.a)):
            if self.a[i]==t:
                return f"{t} found at {i} index"
            else:
                return f"{t} not Found"
            
a = [2,5,3,21,0]
t = 21
l1 = LinearSearch(a,t)
l1.linearsearch(a,t)
```

# Binary Search

```python
class BinarySearch:
    def __init__(self,a ,t):
        self.a = sorted(a)
        self.t = t
        
    def binarysearch(self,a,t):
        if len(self.a)==1:
            return self.a
        else:
            l = 0
            r = len(self.a)-1
            while l<=r:
                m = (l+r)//2
                if self.a[m]==t:
                    return f"{t} found at {m} in the sorted array"
                elif self.a[m]<t:
                    l=m+1
                elif self.a[m]>t:
                    r=m-1
                    
a = [2,5,3,21,0]
t = 21          
b1 = BinarySearch(a,t)
b1.binarysearch(a,t)
```

```python
class BinarySearch:
    def __init__(self, a, t):
        self.a = a 
        self.t = t  
        
    def binarysearch(self,a,t):
        sorted_a = sorted((v, i) for i, v in enumerate(self.a))
        l = 0
        r = len(sorted_a) - 1
        while l <= r:
            m = (l + r) // 2
            
            if sorted_a[m][0] == self.t:  
                
                original_index = sorted_a[m][1] 
                
                return f"{self.t} found at original index {original_index}"

            elif sorted_a[m][0] < self.t:
                l = m + 1
            else:
                r = m - 1

        return f"{self.t} not found"


a = [2, 5, 3, 21, 0]
t = 0
b1 = BinarySearch(a, t)
b1.binarysearch(a,t)
```
